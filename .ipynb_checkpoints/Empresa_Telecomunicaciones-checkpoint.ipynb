## Selección de variables

```python
from sklearn.feature_selection import SelectKBest, f_classif
#Se paramos las variables explicatorias(X) de la variable objetivo(y)
X = df_data_processing_scaled.drop('Churn',axis=1)
y = df_data_processing_scaled["Churn"].values

# Crear instancia de SelectKBest con la función de puntuación f_classif
k_best = SelectKBest(f_classif, k=2)

# Ajustar SelectKBest a los datos de entrenamiento y transformar los datos de entrenamiento
X_new = k_best.fit_transform(X, y)
# X_new tiene la dimensionalidad reducida pero se pierte un pequeño porcentaje de informacion, para este
# caso no lo voy a tener en cuenta

# Obtener los valores p para cada característica
p_values = k_best.pvalues_

feature = X.columns
# Imprimir los valores p
for i in range(len(p_values)):
    print(p_values[i]," Variable: ", feature[i] )
```

Note que:

Por lo que el genero no es significativo y es acorde como lo habiamos dicho en el analisis descriptivo ya que tiene poca significancia. Tambien note:  
0.3269552813531713  Variable:  PhoneService_No  
0.32695528135474317  Variable:  PhoneService_Yes  
Al igual que esta variable de si tiene servicio telefonico.
El resto son significativas y podemos seguir para construir los modelos

Con lo anterior tengo un criterio suficiente para eliminar estos features:

```python
X.drop(['gender_Female','gender_Male','PhoneService_No','PhoneService_Yes'],axis=1,inplace=True)
```

```python
X.shape
```

## Entrenamiento del modelo de regresion logistica binomial

Separamos una base de entrenamiento tomando el 70% de los datos para entrenar el modelo y posteriormete la evaluaremos el modelo con los datos restantes que se separaron, es decir con la base para hacer el test.
separando:

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
#Y listo
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

Listo, ya tenemos nuestro modelo entrenado y ahora nos queda ver que tan bien es capaz de hacer prediciones, es decir si suponemos que tenemos nuevos clientes en la empresa, predecir si se van a reterirar o no antes de que lo hagan, por eso estos algoritmos de machine learning toman valor en las empresas.

Evaluemos la capacidad preditiva con la base de datos de prueba

```python
prediciones = model.predict(X_test)
prediciones
```

```python
valores_reales = y_test
valores_reales
```

```python
from sklearn import metrics
print(metrics.accuracy_score(valores_reales,prediciones))
```

El modelo con Regresion logistica predijo correctamente casi el 80%  de las personas si hacian churn o no

## Entrenamiento del modelo con K-vecinos más cercanos

```python
from sklearn.neighbors import KNeighborsClassifier
def mejor_k(X_test,y_test):
    k_range = range(1, 20)
    scores = []
    for k in k_range:
    	knn = KNeighborsClassifier(n_neighbors = k)
    	knn.fit(X_train, y_train)
    	scores.append(knn.score(X_test, y_test))
    return(k_range[scores.index(max(scores))])
        
mejor_k(X_test,y_test)    
```

Note que el valor de K que nos da mayor presición en el modelo es 18

```python
knn = KNeighborsClassifier(17)
knn.fit(X_train,y_train)
```

```python
prediciones2 = knn.predict(X_test)
prediciones2
```

```python
metrics.accuracy_score(y_test,prediciones2)
```

El modelo con el algoritmo K vecinos más cercanos predijo bien casi el 77% de los casos si las personas hacian churn o no

## Arboles de clasificación y Bagging

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

# Definir el modelo de clasificación base
clf = DecisionTreeClassifier()

# Definir el modelo de Bagging
bagging_clf = BaggingClassifier(base_estimator=clf, n_estimators=10, random_state=0)

# Entrenar el modelo de Bagging
bagging_clf.fit(X_train, y_train)

# Hacer predicciones con el modelo de Bagging
y_pred = bagging_clf.predict(X_test)

# Evaluar la precisión del modelo de Bagging
accuracy = metrics.accuracy_score(y_test, y_pred)
print("Precisión del modelo de Bagging:", accuracy)
```

Este modelo predijo de forma correcta el 76.2% de si las personas hacian Chur o no

## Selección del mejor modelo
Note que la regresion logistica tuvo mayor porcentaje de aciertos de si las personas hacian churn o no por lo que este seria el producto que usaría para clasificar a los nuevos clientes en la empresa y predecir quienes cancelan el servicio y quienes no, ademas este proceso permitió entender cuales son las variables más influyentes como por ejemplo el tipo de contrato (los que pagan mes a mes, 1 año de contrato y 2 años de contrato) el cual es informacion valiosa para aumentar la productividad en la empresa.
