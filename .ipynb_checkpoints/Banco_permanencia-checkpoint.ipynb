### Normalización de los datos continuos
Esto se realiza para mejorar el rendimiento de los algoritmos de aprendizaje automático, y en particular, para evitar que alguna característica tenga mayor peso en el modelo debido a su escala o magnitud.

```python
datos_continuos = datos[["CreditScore","Age","Balance","EstimatedSalary"]]
datos.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_data_processing_scaled = scaler.fit_transform(datos_continuos)
df_data_processing_scaled = pd.DataFrame(df_data_processing_scaled)
df_data_processing_scaled.columns = datos_continuos.columns


# Realizar la combinación utilizando el método merge
df_data_processing_scaled = pd.merge(df_data_processing_scaled, datos[['Geography', 'Gender', 'Tenure','NumOfProducts', 'HasCrCard', 'IsActiveMember','Exited']],
                                    left_index=True, right_index=True)
```

```python
df_data_processing_scaled.head(3)
```

Solo nos queda aplicar los algoritmos.  
Pero antes hagamos selección de variables, hay dos métodos, usar algoritmos de filtro o de envoltura, ambas con sus ventajas y desventajas, para este caso usaré un algoritmo de filtro

## Selección de variables: Algoritmo de filtrado para selección de características

El f_classif es una función de selección de características de sklearn que se utiliza para seleccionar las mejores características para un modelo de clasificación a partir de un conjunto de características. Esta función calcula la relación entre cada característica y la variable objetivo utilizando el análisis de varianza (ANOVA) y devuelve la puntuación F y el valor p correspondiente. esto ya lo he mensionado en anteriores proyectos.

```python
from sklearn.feature_selection import SelectKBest, f_classif

X = df_data_processing_scaled.drop('Exited',axis=1)
y = df_data_processing_scaled["Exited"].values

k_best = SelectKBest(f_classif, k=2)
X_new = k_best.fit_transform(X, y)

p_values = k_best.pvalues_
feature = X.columns

for i in range(len(p_values)):
    print(p_values[i]," Variable: ", feature[i] )
```

Esto nos deja ver un poco el efecto que tiene las variables predictoras sobre la respuesta, sospechó que las variables __HasCrCard__, __Tenure__ y __EstimatedSalary__ pueden que no sean significativas para los modelos, pero por ahora no consideró eliminarlas, cuando monte los modelos observaré la importancia de las variables y tendré todo esto en cuenta.

#### Balanceo
Estaba olvidando hacer el balanceo xD, pero listo, como ya dije, voy a aplicar undersamplig para balancear los datos, para eso voy utilizar el metodo RandomUnderSampler de la libreria imblear.

```python
from imblearn.under_sampling import RandomUnderSampler
undersample = RandomUnderSampler(random_state=42)
X_under, y_under = undersample.fit_resample(X,y) # ACTUALIZO LA (X) Y (y) PERO BALANCEADAS
```

```python
y_under=pd.DataFrame(y_under).rename(columns={0:'Exited'})
```

```python
y_under.value_counts("Exited").reset_index(name="Conteo").pipe(
lambda df: sns.barplot(data=df,x="Exited",y="Conteo",palette = "husl"))
```

```python
y_under[y_under["Exited"]==1].size
```

```python
y_under[y_under["Exited"]==0].size
```

```python
X_under.shape[0]
```

Listo, con esto ya nuestros modelos de machine learning no van a estar sesgados para una categoría, en las últimas líneas de código corroboré que todo estuviera funcionando a la perfección, ya que el número de datos tanto y como de X deben ser las mismas correspondientes a su respectivo índice.

##### Fin del preprocesamiento

#### Función para graficar matrices de confusión:

```python
def graf_matriz_con(y_true, y_pred):
    from sklearn.metrics import confusion_matrix

    # Calcular matriz de confusión
    conf_mat = confusion_matrix(y_true, y_pred)

    # Crear dataframe de la matriz de confusión
    #conf_mat_df = pd.DataFrame(conf_mat,
                               #index=['Actual 0', 'Actual 1'],
                               #columns=['Predicted 0', 'Predicted 1'])
    
    # Trazar la matriz de confusión con heatmap()
    sns.heatmap(conf_mat, annot=True, cmap="coolwarm",fmt='d')
    plt.title('Matriz de confusión')
    plt.xlabel('Etiquetas predichas')
    plt.ylabel('Etiquetas reales')
    plt.show()
```

#### Partiendo la base de datos para entrenamiento y prueba:

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_under,y_under.values.ravel(),test_size=0.2,random_state=42) 
```

## Algoritmo de regresión logistica

```python
from sklearn.linear_model import LogisticRegression
model_LR1 = LogisticRegression()
model_LR1.fit(X_train,y_train)
prediciones = model_LR1.predict(X_test)
```

##### Evaluación:

```python
graf_matriz_con(y_test,prediciones)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,prediciones)
```

Con los datos nuevos tuvo una precisión del 70% para predecir quien salia del banco y quien no en los datos nuevos (Los que se apartaron en la prueba), este resultado tal vez es más o menos bueno dependiendo de la situación.

#### Validacion cruzada

```python
from sklearn.model_selection import cross_val_score
model_LR1 = LogisticRegression()
print(cross_val_score(model_LR1,X_under,y_under.values.ravel(),cv=5))
```

Haciendo validacion cruzada se ve que la precisión se mantiene entre el 71% y 67%

## Entrenamiento del modelo con K-vecinos más cercanos

Para este entrenamiento de k-vecinos más cercanos escojo un n de entre 1 y 20 para ver cuál n me genera la mayor precisión en las predicciones, es importante tener en cuenta que si n es muy grande el modelo puede afectarse e incurrir en overfitting a demás de que computacionalmente sería muy costos.

```python
from sklearn.neighbors import KNeighborsClassifier
def mejor_k(X_test,y_test):
    k_range = range(1, 20)
    scores = []
    for k in k_range:
    	knn = KNeighborsClassifier(n_neighbors = k)
    	knn.fit(X_train, y_train)
    	scores.append(knn.score(X_test, y_test))
    return(k_range[scores.index(max(scores))])
        
mejor_k(X_test,y_test)  
```

```python
model_knn = KNeighborsClassifier(19)
model_knn.fit(X_train,y_train)
```

#### Evaluación

```python
prediciones2 = model_knn.predict(X_test)
```

```python
graf_matriz_con(y_test,prediciones2)
```

##### Evaluación tradicional

```python
accuracy_score(y_test,prediciones2)
```

#### Validación cruzada

```python
model_knn = KNeighborsClassifier()

print(cross_val_score(model_knn,X_under,y_under.values.ravel(),cv=5))
```

Haciendo validación cruzada se ve que la precisión se mantiene entre el 66% y el 71%, muy similar al anterior modelo.

## Algorimo de random Forest

```python
from sklearn.ensemble import RandomForestClassifier
arboles = RandomForestClassifier(n_estimators=100,
                                criterion="gini",
                                max_features="sqrt",
                                bootstrap=True,
                                max_samples=2/3,
                                oob_score=True,
                                random_state=0)

arboles.fit(X_train,y_train)
prediciones = arboles.predict(X_test)
```

```python
graf_matriz_con(y_test,prediciones)
```

#### Evaluación tradicional

```python
accuracy_score(y_test,prediciones)
```

Con los datos nuevos tuvo una precisión del 75% para predecir quien salia del banco y quien no en los datos nuevos (Los que se apartaron en la prueba)

```python
arboles = RandomForestClassifier(n_estimators=100,
                                criterion="gini",
                                max_features="sqrt",
                                bootstrap=True,
                                max_samples=2/3,
                                oob_score=True
                                )
cross_val_score(arboles,X_under,y_under.values.ravel(),cv=5)
```

Haciendo validacion cruzada observamos que la precisión se mantiene entre un 78% y un 75% por lo que este sería el modelo que escogería

```python
arboles.fit(X_train,y_train)
importances = arboles.feature_importances_
```

## Importancia de la característica basada en la disminución media de la impureza (MDI)

```python
feature_name = X_under.columns
```

```python
std = np.std([tree.feature_importances_ for tree in arboles.estimators_], axis=0)
forest_importances = pd.Series(importances, index=feature_name)

fig, ax = plt.subplots()
forest_importances.plot.bar(yerr=std, ax=ax)
ax.set_title("Características importantes usando MDI")
ax.set_ylabel("Disminución media de la impureza")
fig.tight_layout()
```

## Importancia de la característica basada en la permutación de características

```python
from sklearn.inspection import permutation_importance

result = permutation_importance(
    arboles, X_test, y_test, n_repeats=10, random_state=42, n_jobs=2
)

forest_importances = pd.Series(result.importances_mean, index=feature_name)
```

```python
fig, ax = plt.subplots()
forest_importances.plot.bar(yerr=result.importances_std, ax=ax)
ax.set_title("Características importantes usando permutación en el modelo completo")
ax.set_ylabel("Disminución de la precisión media")
fig.tight_layout()
plt.show()
```
